```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from random import randrange



df = pd.DataFrame(columns = ["Arrival Time", "Placement Time 1", "Removal Time 1", "Platelets Extraction", "Placement Time 2", "Removal Time 2", "Plasma Extraction", "Total Service Time"])
technicians = []
centrifuge_platelets = 300
centrifuge_plasma = 480

df["Placement Time 1"] = abs(np.random.uniform(low = 3, high = 6, size = 20))
df["Removal Time 1"] = abs(np.random.uniform(low = 3, high = 6, size = 20))

df["Placement Time 2"] = abs(np.random.uniform(low = 3, high = 6, size = 20))
df["Removal Time 2"] = abs(np.random.uniform(low = 3, high = 6, size = 20))

j = 0
k = 0

for i in range(20):
    technicians.append(0)

def random():
    x = randrange(20)
    return x
    
for x in range(20):
    while (j < 30 or j > 50):
        j = np.random.normal(loc = 40, scale = 5, size = 1)
    while (k < 30 or k > 50):
        k = np.random.normal(loc = 40, scale = 5, size = 1)
    df.loc[x, "Platelets Extraction"] = j
    df.loc[x, "Plasma Extraction"] = k
    j = 0
    k = 0

    
for i in range(20):
    if i == 0:
        df.loc[i, "Arrival Time"] = 0
        df.loc[i, "Total Service Time"] = df.loc[i, "Placement Time 1"] + df.loc[i, "Removal Time 1"] + df.loc[i, "Platelets Extraction"] + df.loc[i, "Placement Time 2"] + df.loc[i, "Removal Time 2"] + df.loc[i, "Plasma Extraction"] 
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Placement Time 1"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Removal Time 1"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Platelets Extraction"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Placement Time 2"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Removal Time 2"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Plasma Extraction"]
    else:
        df.loc[i, "Arrival Time"] = df.loc[i - 1, "Total Service Time"]
        df.loc[i, "Total Service Time"] = df.loc[i, "Placement Time 1"] + df.loc[i, "Removal Time 1"] + df.loc[i, "Platelets Extraction"] + df.loc[i, "Placement Time 2"] + df.loc[i, "Removal Time 2"] + df.loc[i, "Plasma Extraction"] 
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Placement Time 1"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Removal Time 1"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Platelets Extraction"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Placement Time 2"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Removal Time 2"]
        s = random()
        technicians[s] = technicians[s] + df.loc[i, "Plasma Extraction"]

        
time_elapsed = sum(technicians) + centrifuge_platelets + centrifuge_plasma
        
print("Time elapsed since arrival " + str(time_elapsed / 60) + " minutes")
print("Time taken by each indiviual technician for performing activities ")
print(technicians)
print(df)

    
```
